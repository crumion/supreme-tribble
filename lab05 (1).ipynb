## Activity 1: Download

```python
def wget(url):
    ''' Download file from specified url and store in file '''
    response = requests.get(url)
    path     = os.path.basename(url)
    with open(path, 'wb') as fh:
        fh.write(response.content)
```

## Activity 2: Words

```python
def load_words_list(path):
    ''' Load words in file into a list. '''
    wordbank = []
    fh = open(path, 'r')
    for word in fh:
        word = word.lower().strip()
        wordbank.append(word)
    fh = fh.close()
    return wordbank
```

```python
WordsList = load_words_list('words.txt')
```

```python
def load_words_dict(path):
    ''' Loads words in file into a dict. '''
    wordfile = {}
    fh = open(path, 'r')
    for index, word in enumerate(fh):
        word = word.lower().strip()
        wordfile[word] = index
        
    fh.close()
    
    return wordfile
```

```python
WordsDict = load_words_dict('words.txt')
```

## Activity 3: Spell Check

```python
PUNCTUATION = str.maketrans({key: None for key in string.punctuation})

def spell_check_line(line, words):
    ''' Check if each word in line is misspelled by searching words data structure. '''
    line = line.lower().translate(PUNCTUATION).split()
    for word in line:
        if not word in words:
            print('Unknown word: {}'.format(word))
pass
```

```python
spell_check_line('Yayaya, understanding the digital world!', WordsList)
```

```python
spell_check_line('Yayaya, understanding the digital world!', WordsDict)
```

## Activity 4: Benchmark

```python
def spell_check_file(path, words):
    ''' Check if each word in file is misspelled by searching words data structure. '''
    file = open(path, 'r', encoding='utf-8')
    for word in file:
        spell_check_line(word, words)
    pass
```
